# 라인플롯

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

```python
plt.title('라인 플롯')
plt.plot([1, 4, 9, 16])    # y = (x + 1) ** 2
plt.grid()
plt.show()
```

```python
plt.title('라인 플롯')
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 2
plt.plot(xs, ys)    # y = (x + 1) ** 2
plt.grid()
plt.show()
```

### 스타일 지정

```python
# 색깔(r), 마커(s), 선스타일(--)
plt.title("'rs--' 스타일의 plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.grid()
plt.show()
```

```python
plt.title("스타일 적용 예시")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", 
        lw=5, ls='--', marker='o', ms=15)
plt.grid()
plt.show()
```

### 그림 범위 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

### 틱 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid()
plt.show()
```

### 여러개의 선을 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
# t, t, 'r--',/// 직선 /// t, 0.5 * t**2,/// 이차식/// 'bs:', t, 0.2 * t**3, 'g^-'///3차식///
plt.grid()
plt.show()
```

```python
# plt.plot() 을 많이 써도 상관없다
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

### 겹쳐그리기

```python
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
# plt.hold(True)   # <- 1,5 버전에서는 이 코드가 필요하다.
plt.plot([9, 16, 4, 1],
         c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c")
# plt.hold(False)  # <- 1,5 버전에서는 이 코드가 필요하다.
plt.show()
```

### 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend()                            # loc=2
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend(loc=8)                            # loc='lower center'
plt.grid()
plt.show()
```

### x축, y축, 라벨 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.grid()
plt.show()
```
