# 여러개의 그림 그리기

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

### Figure 객체

```python
np.random.seed(2021)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

### Axes객체와 subplot

```python
# 2 * 1 행이 두개 열이 하나, 그 뒤에 오는 숫자는 몇번째 그래프를 지칭하는지!

subplot(2, 1, 1)
# 여기에서 윗부분에 그릴 플롯 명령 실행
subplot(2, 1, 2)
# 여기에서 아랫부분에 그릴 플롯 명령 실행
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)                  # Axes
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2)                  # Axes
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)

plt.subplot(221)                # 행의 갯수, 열의 갯수, 순서
plt.plot(np.random.rand(5))
plt.title("axes 1")
plt.grid()

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")
plt.grid()

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")
plt.grid()

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")
plt.grid()

plt.tight_layout()
plt.show()
```

- subplots()

```python
fig, axes = plt.subplots(2, 2)          # 2 x 2 array

np.random.seed(2021)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 0].grid()
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[0, 1].grid()
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 0].grid()
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")
axes[1, 1].grid()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)          # 2 x 2 array
np.random.seed(2021)

for i in range(4):
    ax = axes[i//2, i%2]                # [행 , 열]
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes {i+1}')
    ax.grid()


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10,4))          # 2 x 2 array
np.random.seed(2021)

for i in range(2):
    ax = axes[i]                # [행 , 열]
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes {i+1}')
    ax.set_xlabel('X'); ax.set_ylabel('Y')
    ax.grid()

fig.suptitle('2개 그림', fontsize=12)
plt.tight_layout()
plt.show()
```
