# Matplotlib의 다양한 플롯

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

## 1. Bar Chart
- x 데이터가 카테고리 값인 경우


```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(2021)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))                     # 데이터의 신뢰구간(여기서는 랜덤이라 의미 x)

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)     # alpha=투명도
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.grid()
plt.show()
```

### 타이타닉 선실별 승객수

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
s = titanic['class'].value_counts()
s
```

```python
df = titanic[['class', 'pclass']].groupby('class').count()
df
```

```python
y = s.values
x = np.arange(len(y))
xlabel = s.index.tolist()

plt.title('타이타닉 선실별 승객수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel('선실')
plt.ylabel('승객수')
plt.show()
```

```python
y = df.pclass.values
x = np.arange(len(y))
xlabel = df.index.tolist()

plt.title('타이타닉 선실별 승객수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel('선실')
plt.ylabel('승객수', rotation=0, labelpad=15)
plt.show()
```

## 2. 파이 차트
- 카테고리 별 값의 상대적인 비교를 할 때
- plt.axis('equal')

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]                                        # 합이 100
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart", fontsize=15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)          # 90도로 해두어야 12시에서 시작
plt.axis('equal')
plt.show()
```

## 6. Box Plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

- Setosa 품종의 각 feature의 박스 플롯

```python
setosa = iris[iris.species == 'setosa']
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length, setosa.petal_width],
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

plt.title('Setosa 품종의 각 feature별 각 박스 플롯')
plt.show()
```

- 모든 품종의 피쳐별 박스 플롯

```python
fig, axes = plt.subplots(1, 3, figsize=(12,8), sharey=True)

for i, species in enumerate(iris.species.unique()):     # enumerate - 값과 키 모두
    sub = iris[iris.species == species]
    ax = axes[i]
    ax.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length, setosa.petal_width],
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    ax.set_title(species)
    ax.grid()

fig.suptitle("모든 품종의 피쳐별 박스 플롯", fontsize=16, y=1.03)
plt.tight_layout()
plt.show()
```
